```python
#seprating data and label
x=orignal_dibates_dataset.drop(columns='Outcome' ,axis=1)
y=orignal_dibates_dataset['Outcome']
```

```python
print(x)
print(y)
```

Data standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardize_data=scaler.transform(x)
```

```python
print(standardize_data)
```

```python
x=standardize_data
y=orignal_dibates_dataset['Outcome'] 
```

```python
print(x)
print(y)
```

DATA SPLITING

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify= y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

MODEL TRAINNIG

```python
#classifier=svm.SVC(kernel='linear')
classifier=KNeighborsClassifier(n_neighbors=5)
```

```python
classifier.fit(x_train,y_train)
```

MODEL EVALUATION


ACCURACY SCORE

```python
#accuracy score on the training data
x_train_pred=classifier.predict(x_train)
x_train_pred_accuracy=accuracy_score(y_train,x_train_pred)
```

```python
print(x_train_pred_accuracy)
```

```python
# accuracy on test data
x_test_pred=classifier.predict(x_test)
x_test_pred_accuracy=accuracy_score(y_test,x_test_pred)
```

```python
print(x_test_pred_accuracy)
```

MAKING A PREDICTIVE SYSYTEM

```python
input_data=(11,120,80,37,150,42.3,0.785,48)


#changing input_data to numpy array
input_data_as_numpy_array=np.asarray(input_data)

# reshaping the array as we are we predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)


# standardize the input data
std_data=scaler.transform(input_data_reshaped)
#print(std_data)

prediction=classifier.predict(std_data)
#print(prediction)
```

```python
if prediction==0:
  print("NOT DIBATIC")
else:
  print("DIBATIC")
```
